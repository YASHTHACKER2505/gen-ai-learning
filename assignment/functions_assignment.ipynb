### Task 1 – Basic Function: Price After Discount

```python
def apply_discount(price, discount=5):
    if discount > 60:
        discount = 60  
    discount_amount = price * (discount / 100)
    last_price=price - discount_amount
    return last_price
```

```python
print(apply_discount(1000, 10))
print(apply_discount(500))  
```

### Task 2 – Recursive Function: Factorial Utility

```python
def factorial(n):
    if n < 0:
        print("Negative numbers are not allowed")
        return None
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
```

```python
print(factorial(5))
print(factorial(0))
print(factorial(-3))
```

### Task 3 – Lambda Function: GST Calculator

```python
gst = lambda price: price + (0.18 * price)
```

```python
print(gst(100))
```

### Task 4 – Using map(): Apply GST to List of Prices

```python
prices = [100, 250, 400, 1200, 50]
```

```python
prices_with_gst = list(map(gst, prices))
```

```python
print("Original Prices:", prices)
print("Prices after GST:", prices_with_gst)
```

### Task 5 – Using filter(): Filter Expensive Products

```python
prices = [100, 250, 400, 1200, 50, 2000, 850]
```

```python
grt_500 = list(filter(lambda x: x > 500, prices))
```

```python
les_eq_500 = list(filter(lambda x: x <= 500, prices))
```

```python
print("Prices > 500:", grt_500)
print("Prices <= 500:", les_eq_500)
```

### Task 6 – Combined Utility Function

```python
def process_prices(prices):
    discounted_prices = list(map(lambda x: x - (0.10 * x), prices))
    filtered_prices = list(filter(lambda x: x > 300, discounted_prices))
    return discounted_prices, filtered_prices
```

```python
result = process_prices([100, 500, 900, 50, 750])
```

```python
print("Discounted Prices:", result[0])
print("Filtered Prices :", result[1])
```

### Task 7 – Mini Problem: Menu Using Functions

```python
def add_price(prices_list, price):
    prices_list.append(price)
```

```python
def get_average_price(prices_list):
    return sum(prices_list) / len(prices_list) if prices_list else 0
```

```python
def get_max_price(prices_list):
    return max(prices_list) if prices_list else 0

```

```python
prices = []
```

```python
while True:
    print("\nMenu")
    print("1 - Add price")
    print("2 - Show average price")
    print("3 - Show highest price")
    print("q - Quit")

    choice = input("Enter choice: ")

    if choice == "1":
        price = float(input("Enter price: "))
        add_price(prices, price)

    elif choice == "2":
        print("Average price:", get_average_price(prices))

    elif choice == "3":
        print("Highest price:", get_max_price(prices))

    elif choice == "q":
        print("Exited")
        break

    else:
        print("Invalid choice")
```

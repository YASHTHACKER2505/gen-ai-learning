### TASK 3 Product Pricing (Dictionaries)

```python
price_dict = {
    "Laptop": 120000,
    "Tv": 25000,
    "Printer": 5000,
    "Mouse": 3000,
    "Keyboard": 1500,
    "Speaker": 800,
    "Monitor":12000
}
```

```python
price_dict["Headphones"] = 4500
```

```python
price_dict["Mouse"] = 4000
```

```python
print("updated dictionary : ",price_dict)
```

```python
remove_product = "Tablet"
if remove_product in price_dict:
    del price_dict[remove_product]
else:
    print(remove_product, "not found in price list")
```

```python
total_price = sum(price_dict.values())
avg = total_price / len(price_dict)
print("Average product price:", avg)
```

```python
max_product = max(price_dict, key=price_dict.get)
min_product = min(price_dict, key=price_dict.get)
print("max:", max_product, price_dict[max_product])
print("min:", min_product, price_dict[min_product])
```

### Task 4: Combined Operations

```python
catalog = []
print(products)
print(categories)
print(price_dict)
```

```python
for i in range(len(products)):
    product_name = products[i]
    price = price_dict[product_name]
    category = categories[i]
    catalog.append((product_name, price, category))
```

```python
category_to_products = {}
```

```python
for i in catalog:
    category = i[2]
    product = i[0]

    if category not in category_to_products:
        category_to_products[category] = []

    category_to_products[category].append(product)
print(category_to_products)
```

```python
maxc= ""
count = 0

for i in category_to_products:
    temp = len(category_to_products[i])

    if temp > count:
        count = temp
        maxc = i
        print(maxc)

print("Category with most products:", maxc)
```

```python
print("products are :")
for i in category_to_products[maxc]:
    print(i)
```

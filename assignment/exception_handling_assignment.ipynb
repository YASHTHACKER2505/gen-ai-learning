### Task 1: Safe Division Utility

```python
try:
    num = float(input("Enter numerator: "))
    den = float(input("Enter denominator: "))
    result = num / den
except ValueError:
    print("Invalid input. Please enter numbers only.")
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
else:
    print("Result:", result)
finally:
    print("Operation Complete")
```

### Task 2: Bill Calculator with Error Handling

```python
prices = [120, 350, 'abc', 500, -200, 800]
```

```python
total = 0
```

```python
for price in prices:
    try:
        if not isinstance(price, (int, float)):
            raise TypeError("Not a number")

        if price < 0:
            raise ValueError("Negative price not allowed")

        total += price
        print("Running total:", total)

    except TypeError:
        print("Invalid type skipped:", price)
    except ValueError as e:
        print(e)
```

### Task 3: Custom Exception – Age Validator

```python
def check_age(age):
    if age < 1 or age > 120:
        raise ValueError("Age must be between 1 and 120")

```

```python
try:
    age = int(input("Enter your age: "))
    check_age(age)
    print("Age is valid.")
except ValueError as e:
    print(e)
```

### Task 4: File Reader with Exception Handling

```python
filename = input("Enter filename: ")
```

```python
try:
    file = open(filename, "r")
    lines = file.readlines()
    for line in lines[:3]:
        print(line.strip())
except FileNotFoundError:
    print("File not found.")
except PermissionError:
    print("Permission denied.")
finally:
    print("File operation attempted.")
```

### Task 5: Mini Program – Safe Shopping Cart

```python
cart = []
```

```python
while True:
    user_input = input("Enter price (or 'q' to quit): ")

    if user_input.lower() == 'q':
        break

    try:
        price = float(user_input)

        if price < 0:
            raise ValueError("Negative price not allowed")

        cart.append(price)

    except ValueError as e:
        print(e)
```

```python
print("Total items:", len(cart))
print("Total bill:", sum(cart))
```

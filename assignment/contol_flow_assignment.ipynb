### Task 1: Discount Rules (if / elif / else)

```python
user_input = input("Enter order amount: ")
try:
    order_amount = int(user_input)
except:
    print("Error: Please enter a valid numeric order amount.")
```

```python
if order_amount >= 2000:
    discount = order_amount * 15/100
elif order_amount >= 1500:
    discount = order_amount * 10/100
elif order_amount >= 1000:
    discount = order_amount * 7/100
else:
    discount = order_amount * 0/100
```

```python
final_amount = order_amount - discount
```

```python
print("Final Amount:", final_amount)
```

```python
tax = final_amount * 5/100
total = final_amount + tax
```

```python
print("Final_amount:",final_amount)
print("tax :",tax)
print("total :",total)
```

### Task 2: Process Multiple Orders (for loop)

```python
orders = [1200, 2500, 800, 1750, 3000]
data = []
```

```python
total_revenue = 0
get_discount = 0
```

```python
for order in orders:
    if order >= 2000:
        discount = 15
    elif order >= 1500:
        discount = 10
    elif order >= 1000:
        discount = 7
    else:
        discount = 0

    final_amount = order - (order * discount / 100)
    total_revenue += final_amount

    if discount > 0:
        get_discount += 1
    data.append([order, f"{discount}%", final_amount])
```

```python
print("Order | Discount % | Final Amount")
print("---------------------------------")
for i in data:
    print(i[0], "|",i[1] ,"|", i[2])
print("---------------------------------")
```

```python
print("Total Revenue After Discounts:", total_revenue)
print("Orders with Discount:", get_discount)
```

### Task 3: User Menu (while loop + break / continue)

```python
orders = []
```

```python
while True:
    print("\nMenu:")
    print("1 - Add order amount")
    print("2 - Show all orders with discounts")
    print("q - Quit")
    choice = input("Choose an option: ")

    if choice == "1":
        amount_input = input("Enter order amount: ")
        try:
            amount = int(amount_input)
            orders.append(amount)
        except:
            print("Invalid amount. Try again.")
            continue

    elif choice == "2":
        if len(orders) == 0:
            print("No orders available.")
            continue

        for order in orders:
            if order >= 2000:
                discount = 15
            elif order >= 1500:
                discount = 10
            elif order >= 1000:
                discount = 7
            else:
                discount = 0

            final_amount = order - (order * discount/100)
            print("Order:", order, "\nFinal:", final_amount)

    elif choice == "q":
        print("Program Exited")
        break

    else:
        print("Invalid option. Try again.")
        continue
```

### Task 4: Loop Control with Conditions (break & continue)

```python
daily = [200, 150, 0, 400, 50, -1, 300]
```

```python
total_sales = 0
```

```python
for i in daily:
    if i == -1:
        break

    if i == 0:
        continue

    total_sales += i
```

```python
print("Total Sales:", total_sales)
```

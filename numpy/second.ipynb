```python
##STACKING AND SPLITTING->SIZE OF THE ARRAY SHOULD SAME
#ARRAY1+ARRAY2=STACKING(hstack((args)),vstack((args)))
#SPLITING:SPLIT THE ARRAY(hsplit(array,bhag),vsplit(array,bhag)
#horizontal->row
#vertical->column
```

```python
a1=np.arange(1,13).reshape(3,4)
a1
```

```python
a2=np.arange(13,25).reshape(3,4)
a2
```

```python
h_a=np.hstack((a1,a2))
```

```python
h_a
```

```python
v_a=np.vstack((a1,a2))
v_a
```

```python
h_s=np.hsplit(a1,2)
h_s
```

```python
v_s=np.vsplit(v_a,2)
v_s
```
